#### Some web scraping has been done to get a lot of words or expressions related to skin cancer. Essentially we took all the link from "Melanoma"  wikipedia url, because we've seen that all the links in blue were about the main topic we had. We selected randomly 200 queries with the same probability of taking them from the top or the bottom of the selected url.
####  Once we had all the keywords, we used  pytrends package to request proxy data we'd needed it. Some other keywords have been added manually, since they had a good distribution of data over all states of USA.

```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import numpy.random
import time

cnt = requests.get('https://en.wikipedia.org/wiki/Melanoma')
cnt = cnt.text.split('<th scope="row" class="navbox-group" style="width:1%"><a href="/wiki/Help:Authority_control" title="Help:Authority control">Authority control</a></th>')[0] 
soup = BeautifulSoup(cnt, 'lxml')
kws = []
for row in soup.find_all('a'):
    if str(row.get('title'))!='None' and str(row.text)!='edit':
        kws += [str(row.get('title'))+'---'+str(row.text)]

numpy.random.seed(123)
kws = [kw.split('---')[0] for kw in kws[5:] if 'Identifier' not in kw]
kws = list(set(kws))
kws = numpy.random.permutation(kws)[:200]
```

```python
pytrends = TrendReq(hl='en-US', tz=360)

kw_list = ['Skin cancer','Skin cancer symptoms','Carcinoma',
           'Melanoma','Chemotherapy','Cancer','Basalioma',
           'Melanoma skin cancer','Mole cancer', 'Skin cancer types', 
           'Cancerous moles', 'Melanocytes', 'Skin melanoma', 'Skin sun', 
           'Radiation cancer', 'Prednisone', 'Ionizing radiation', 
           'Causes of skin cancer', 'What causes skin cancer', 
           'What are causes of skin cancer', 'Basal cell skin cancer', 'Merkel cell carcinoma', 
           'Skin Care','uv protection','white spots ','sunscreen','solarium','sunburn']
kw_list += list(kws)
kw_list = list(set(kw_list))

def googleTrends(years,kw_list,cat,states,path):
    writer = pd.ExcelWriter(path+'GoogleTrends/QUERIES(new)-cat'+str(cat)+'.xlsx')
    for year in years:
        print(year)
        db = pd.DataFrame(index = states)
        for kw in kw_list:
            print(kw)
            pytrends.build_payload([kw], cat=cat, timeframe= str(year)+'-01-01 '+str(year)+'-12-31', geo='US', gprop='')
            time.sleep(5)
            try:
                app = pytrends.interest_by_region(resolution='USA')
                #print(app)
                db = pd.merge(db, app, how = 'left', left_index = True, right_index = True)
            except KeyError:
                print('not found')
                
        db.to_excel(writer, str(year), index = True)
    writer.save()
    return(kw_list)
```

#### We considered 4 different categories for each query/keyword : all categories, health , health condition and cancer.

### let's save the keywords found for all years

```python
import pandas as pd
years = list(range(2011,2017))
categories = [0,45,419,429]
path = 'D:/DataScience/DE/HW02/'


for cat in categories:
    for year in range(len(years)):
        df = pd.read_excel(path+'GoogleTrends/QUERIES(new)-cat'+str(cat)+'.xlsx', sheetname = year)
        #print(cat,years[year])
        if year == 0:
            kws = set(df.loc[:,df.isnull().sum()<10].columns).difference(['LocationDesc'])
        kws = kws.intersection(df.loc[:,df.isnull().sum()<10].columns)
    file = open(path+'GoogleTrends/keywords(new)'+str(cat)+'.txt','w')
    s = ''
    for kw in kws: s+= str(kw)+'-'
    file.write(s) 
    file.close()
```
